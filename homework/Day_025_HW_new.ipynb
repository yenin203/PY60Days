```python
# 導入開發套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# 新建地圖
map = Basemap()  #Basemap有很多屬性，這里全都使用默認参數

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()

# 存储结果
plt.savefig('test.png')
```

```python
# 1. merc(Mercator)：麥卡托投影 (航空圖、航海圖、赤道附近地區地圖)

# 2. tmerc(Transverse Mercator)：橫麥卡托投影 (小區域的地形圖)

# 3. npstere,spstere(North(South)-Polar Stereographic)：極投影 (極區地形圖、航海圖)

# 4. lcc(Lambert Conformal：蘭伯特投影 (中緯度地形圖、天氣圖、航空圖)

# 5. robin(Robinson)：羅賓森投影 (一般世界地圖)
```

```python
# 以下設定麥卡托投影為例：
# 新建地圖
map = Basemap(projection='merc',llcrnrlon=-180, urcrnrlon=180, llcrnrlat=-80, urcrnrlat=80)
# 經度 : longitude 、 緯度 : latitude
# llcrnrlon – 左下角的經度。
# urcrnrlon – 右上角的經度。
# llcrnrlat – 左下角的緯度。
# urcrnrlat – 右上角的緯度。


# 畫圖
map.drawcoastlines()

map.drawparallels(range(-90,91,30), color='#CCCCCC')
map.drawmeridians(range(-180,181,60), color='#CCCCCC')

#畫出州。
map.drawstates(color='b') 

# 畫出城市
map.drawcountries(color='darkred')

# 劃出海的顏色
map.drawmapboundary(fill_color='aqua')

# 劃出陸地的顏色
map.fillcontinents(color = 'coral')
# 顯示结果
plt.show()

# 存储结果
plt.savefig('test.png')
```

```python
# 可通過 drawmeridians() 和 drawparallels() 繪製經線和緯線。
# 新建地圖
map = Basemap(projection='merc',urcrnrlat=80, llcrnrlat=-80,llcrnrlon=-180, urcrnrlon=180)  #Basemap有很多屬性，這里全都使用默認参數

# 畫圖
map.drawcoastlines()

map.drawparallels(range(-90,91,30), color='#CCCCCC')  # 緯線
map.drawmeridians(range(-180,181,60), color='#CCCCCC')  # 經線

# 顯示结果
plt.show()

# 存储结果
plt.savefig('test.png')
```

```python
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'aqua')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'coral', lake_color = 'aqua')

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```
