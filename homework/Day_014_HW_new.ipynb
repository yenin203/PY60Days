### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)
df = pd.DataFrame(data)
df
# print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
# df['subscriptionCount_label'] = LabelEncoder().fit_transform(df['subscriptionCount'].values)
result_df = df.sort_values('subscriptionCount', ascending=False)
result_df.to_csv('dcard_famous_data.csv')
```
