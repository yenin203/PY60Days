```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
n = 100
y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.show()
plt.savefig("cos_line.png",dpi=300,format="png")
```

```python
y_sin = np.sin(x)
plt.plot(x, y_sin)
```

```python
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
Y2 = np.random.normal(0, 1, n)
font_set = FontProperties(fname=r"/Users/shijung.huang/Documents/shijung-font/DFT_ZY9.ttc", size=15)
plt.scatter(X, Y, color ='orange')
plt.scatter(X, Y2, color = 'black')
plt.title(u'給定散點圖顏色', fontproperties=font_set)
```

```python
# x, y: float or array-like, shape (n, )
# s: float or array-like, shape (n, ), optional
# The marker size in points**2. Default is rcParams['lines.markersize']
# c: array-like or list of colors or color, optional
# alpha: float, default: None
# The alpha blending value, between 0 (transparent) and 1 (opaque).
xyz=np.array(np.random.random((n, 3)))
plt.scatter(X, Y, s=50, c=xyz, alpha=0.9)
```
