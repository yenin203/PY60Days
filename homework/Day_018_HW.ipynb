```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
x = np.arange(0, 3 * np.pi, 0.1)

y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.show()
plt.savefig("cos_line.png",dpi=300,format="png")
```

```python
y_sin = np.sin(x)
plt.plot(x, y_sin)
```
