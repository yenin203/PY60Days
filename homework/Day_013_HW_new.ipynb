````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
````

````python
pd.merge(df1, df2, on='fruit', how='outer')
````

````python
# - 依照 fruit 欄位做合併
df12 = pd.concat([df1, df2]).reset_index(drop=True)
df12.index.name = "id"
df12
````

### 2. [簡答題] 承上題，請問為什麼依照 merge 合併後有些資料不見了？


````python
# pine只在df2,banana只在df1
# 因為預設 inner 資料看依照左邊還右邊排序如果沒有那個水果就會被濾掉
# 改為outer就有全部
````

### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

````python
df1.index.name = 'id'
df2.index.name = 'id'
pd.merge(df1, df2, on='id',suffixes=('_df1', '_df2'), how='left')
````
